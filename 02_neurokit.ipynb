```python
data = np.load('sample-data/0004.npy')
data = np.squeeze(data)
print(data.ndim)
print(type(data))
print(len(data))
print((data))
```

```python
# Process ecg for 1 Person TODO: for-loop
ecg_signals, info = nk.ecg_process(data, sampling_rate=1000)
plot = nk.ecg_plot(ecg_signals[:30000], sampling_rate=1000)
```
